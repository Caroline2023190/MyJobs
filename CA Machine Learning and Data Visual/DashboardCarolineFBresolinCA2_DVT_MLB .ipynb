```python
def apply_custom_style(fig, title):
    fig.update_layout(
        title=title,
        plot_bgcolor='white',  # Plot background color
        paper_bgcolor="#EEF4F4",  # Paper background color
        font=dict(color='rgb(50, 50, 50)'),  # Font color
        xaxis=dict(
            showgrid=False,  # Remove x-axis grid
            zeroline=False,
            title_font=dict(size=12),
            tickfont=dict(size=12)
        ),
        yaxis=dict(
            showgrid=True,  # Keep y-axis grid
            gridcolor='#F1FFFF',
            title_font=dict(size=12),
            tickfont=dict(size=12)
        ),
        title_x=0.5, # Centralized title
        title_font=dict(size=14)
    )
```

```python
import plotly.graph_objs as go
import pandas as pd
import plotly.express as px
from dash import Dash, dcc, html
from dash.dependencies import Input, Output

app = dash.Dash(__name__, external_stylesheets=[dbc.themes.QUARTZ])

app.layout = html.Div([
    # Header
    html.Div([
        html.H1('Groceries Store Dashboard'),
        html.P('Visualizing key metrics and insights for a Groceries Store')
    ], className='jumbotron text-center'),
    
    # Sidebar
    html.Div([
        html.Ul([
            html.Li(html.A('Overview', href='#overview')),
            html.Li(html.A('Sales Analysis', href='#sales-analysis')),
        ], className='nav flex-column')
    ], className='col-md-2'),
    
    # Main Content
    html.Div([
        # Overview Section
        html.Div([
            html.H2('Overview', id='overview'),

        # Dropdowns and RadioItems
        dcc.Dropdown(
            id='item-description-dropdown',
            options=[{'label': 'All', 'value': 'all'}] + [{'label': item, 'value': item} for item in df['itemDescription'].unique()],
            value='all',
            clearable=False,
            style={'color': 'black'}
        ),
        dcc.Dropdown(
            id='year-dropdown',
            options=[{'label': str(year), 'value': year} for year in df['Year'].unique()],
            value=df['Year'].unique(),  # Select all years by default
            multi=True,
            clearable=False
        ),
        dcc.RadioItems(
            id='chart-type-radio',
            options=[
                {'label': 'Line Chart', 'value': 'line'},
                {'label': 'Bar Chart', 'value': 'bar'}
            ],
            value='line',
            labelStyle={'display': 'inline-block'}
        )
    ], className='row'),
        dcc.Graph(id='quantity-over-months'),
        ], className='col-md-10'),

    # Sales Analysis Section
    html.Div([
        html.H2('Sales Analysis', id='sales-analysis'),
        html.Div([
            dcc.Graph(id='frequency-items'),
            dcc.Graph(id='top-products'),
        ], className='row'),
    ], className='col-md-10'),
])

# Callback to update the sales summary graph
@app.callback(
    Output('frequency-items', 'figure'),
    [Input('frequency-items', 'id')]
)
def update_frequency_items(_):
    Frequency_of_items = df.groupby(pd.Grouper(key = 'itemDescription')).size().reset_index(name = 'count')
    fig = px.treemap(Frequency_of_items, path = ['itemDescription'], values = 'count')
    apply_custom_style(fig, 'Frequency purchased items')
    return fig

# Callback to update the top products graph
@app.callback(
    Output('top-products', 'figure'),
    [Input('top-products', 'id')]
)
def update_top_products(_):
    top_products = df['itemDescription'].value_counts().head(10)
    fig = px.bar(top_products, x=top_products.index, y=top_products.values, labels={'x': 'Item', 'y': 'Quantity'}, color_discrete_sequence=['#2E46A4'])
    apply_custom_style(fig, 'Top 10 Purchased Products')
    return fig
    
# Callback to update quantity over months
@app.callback(
    Output('quantity-over-months', 'figure'),
    [Input('item-description-dropdown', 'value'),
     Input('year-dropdown', 'value'),
     Input('chart-type-radio', 'value')]
)
def update_graph(selected_item, selected_years, chart_type):
    filtered_df = df[df['Year'].isin(selected_years)]

    if selected_item != 'all':
        filtered_df = filtered_df[filtered_df['itemDescription'] == selected_item]
    grouped_df = filtered_df.groupby(['Year', 'Month'])['Quantity'].sum().reset_index()

    colors = ['#11598B', '#FAEC0E']

    fig = go.Figure()
    
    for year, color in zip(grouped_df['Year'].unique(), colors):
        year_df = grouped_df[grouped_df['Year'] == year]
        if chart_type == 'line':
            fig.add_trace(go.Scatter(x=year_df['Month'], y=year_df['Quantity'], mode='lines', name=str(year), marker=dict(color=color)))
        else:
            fig.add_trace(go.Bar(x=year_df['Month'], y=year_df['Quantity'], name=str(year), marker=dict(color=color)))
            
    fig.update_layout(
        title=f'Quantity of {selected_item} Over Months' if selected_item != 'all' else 'Quantity of All Items Over Months',
        xaxis_title='Month',
        yaxis_title='Quantity',
        barmode='group' if chart_type == 'bar' else None,
        plot_bgcolor='white',  # Plot background color
        paper_bgcolor="#EEF4F4",  # Paper background color
        font=dict(color='rgb(50, 50, 50)'),  # Font color
        xaxis=dict(
            showgrid=False,  # Remove x-axis grid
            zeroline=False,
            title_font=dict(size=12),
            tickfont=dict(size=12)
        ),
        yaxis=dict(
            showgrid=True,  # Keep y-axis grid
            gridcolor='#F1FFFF',
            title_font=dict(size=12),
            tickfont=dict(size=12)
        ),
        title_x=0.5, # Centralized title
        title_font=dict(size=14)
    )
    
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
